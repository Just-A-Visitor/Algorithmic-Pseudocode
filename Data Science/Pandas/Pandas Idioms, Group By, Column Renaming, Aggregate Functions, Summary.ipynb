##### Let's read the census data

## Boolean masking with the help of where


1) Where is used to do boolean masking
2) dropna is used to drop rows with na values
3) set_index is used for multi level indexing
4) rename is used to rename columns

```python
#The whole statement should be enclosed in parenthesis, to span multiple lines

(df.where(df['SUMLEV']==50)
    .dropna()
    .set_index(['STNAME', 'CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))

```

```python
#As we can see, only the view was changed not the original document

df.head()
```

# Picking only the tuples which satisfy a given criteria

```python
df = pd.read_csv('census.csv')

df = df[df['SUMLEV']==50]

df.head()
```

# Using Group By

# When you iterate over the group by view, the first iterbale gives you the group leader's name, the second iterable gives you the data frame

```python
for group,frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in State ' + group + ' have an average 2010 census of ' + str(avg))
```
