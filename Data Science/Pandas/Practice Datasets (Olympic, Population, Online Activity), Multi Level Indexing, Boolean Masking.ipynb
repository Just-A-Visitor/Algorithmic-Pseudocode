```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
df.head()
```

# A part that I did not understand

```python
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#' + col[1:]}, inplace=True) 

df.head()
```

## Let's query the countries which hav achieved gold medal at the summer olympics 

# Boolean Mask Demonstration

```python
copy = df['Gold'] > 0
copy.head()
# This broadcasts the comparison operator
```

## Let's merge this boolean mask with the original data frame

# Where

```python
only_gold = df.where(df['Gold']>0)
only_gold.head()
```

## Note that a lot of ENTRIES have NAN values, means they are not part of the answer.
# Counting the answers would consider only the countries with gold

```python
only_gold['Gold'].count()
```

```python
df['Gold'].count()
```

# Let's drop the rows with no data using dropna

```python
only_gold = only_gold.dropna()
only_gold.head()
```

# An easier way to filter rows without using where and overloading with the boolean masking

```python
only_gold = df[df['Gold']>0]
only_gold.head()
```

## Interlinking boolean masks to create more complex masks

```python
df[(df['Gold']>0) | (df['Gold.1']>0)]

# This gives us the list of countries which have won a gold at either the winter or summer olympics
```

### Is there a country which has only one a gold in the winter olympics but never in the summer olympics?

```python
df[(df['Gold']==0) & (df['Gold.1']>0)]
```
